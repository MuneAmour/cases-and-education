```python
#–û—Ç–±–µ—Ä–µ–º 10 —Å–∞–º—ã—Ö –ø—Ä–∏–±—ã–ª—å–Ω—ã—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º –∏ –æ–±–µ—Ä–Ω–µ–º –∏—Ö –≤ —Å–ø–∏—Å–æ–∫
top_10_genres = game_data.pivot_table(index = 'genre', 
                      values = 'sales_total', 
                      aggfunc = 'sum').sort_values(by = 'sales_total', ascending = False)[:10].reset_index()
top_10_genres = top_10_genres['genre']
top_10_genres
```

```python
# –û—Ç—Ñ–∏–ª—å—Ç—Ä—É–µ–º –∑–Ω–∞—á–µ–Ω–∏—è –¥–∞—Ç–∞—Ñ—Ñ—Ä–µ–π–º–∞ –ø–æ 10 —Å–∞–º—ã–º –ø—Ä–∏–±—ã–ª—å–Ω—ã–º –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º –∏ –ø–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫, –æ—Ç—Ä–∞–∂–∞—é—â–∏–π –¥–∏–Ω–∞–º–∏–∫—É –ø—Ä–æ–¥–∞–∂
genre_sales = game_data[(game_data['genre'].isin(top_10_genres) & 
                        (game_data['year_of_release'] >= 2010))].pivot_table(index = ['year_of_release'],
                                                                             columns = 'genre',
                                                                             values = 'sales_total', 
                                                                             aggfunc = 'mean')
                           
genre_sales.plot(grid = True, figsize = (14,6))
plt.legend(loc = 'upper left')
plt.title('–î–∏–Ω–∞–º–∏–∫–∞ –æ–±—ä–µ–º–∞ —Å—Ä–µ–¥–Ω–∏—Ö –ø—Ä–æ–¥–∞–∂ 10 —Å–∞–º—ã—Ö –ø—Ä–∏–±—ã–ª—å–Ω—ã—Ö –∂–∞–Ω—Ä–æ–≤', loc = 'center', pad = 10)
plt.xlabel('–ì–æ–¥')
plt.ylabel('–û–±—â–∏–π –æ–±—ä–µ–º –ø—Ä–æ–¥–∞–∂')
```

```python
# –°–æ–∑–¥–∞–¥–∏–º –ø–µ—Ä–µ–º–µ–Ω—ã–µ, —Ö—Ä–∞–Ω—è—â–∏–µ –¥–∞–Ω–Ω—ã–µ –ø–æ —Ä–∞–∑–Ω—ã–º —Ä–µ–≥–∏–æ–Ω–∞–º 
na_customer = good_data[['year_of_release', 'genre', 'platform','rating_new','na_sales']]
eu_customer = good_data[['year_of_release', 'genre', 'platform','rating_new','eu_sales']]
jp_customer = good_data[['year_of_release', 'genre', 'platform','rating_new','jp_sales']]
other_customer = good_data[['year_of_release', 'genre', 'platform','rating_new','other_sales']]
```

```python
# –û—Ç–±–µ—Ä–µ–º —Ç–æ–ø 5 –ø–ª–∞—Ç—Ñ–æ—Ä–º –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —Ä—ã–Ω–∫–∞ (–Ω–∞ –∞–∫—Ç—É–∞–ª—å–Ω—ã–π –ø–µ—Ä–∏–æ–¥)

# NA region
top_5_platforms_na = na_customer[na_customer['year_of_release'] >= 2013].pivot_table(index = 'platform', 
                      values = 'na_sales', 
                      aggfunc = 'sum').sort_values(by = 'na_sales', ascending = False)[:5].reset_index()
top_5_platforms_na = top_5_platforms_na['platform']


# EU region
top_5_platforms_eu = eu_customer[eu_customer['year_of_release'] >= 2013].pivot_table(index = 'platform', 
                      values = 'eu_sales', 
                      aggfunc = 'sum').sort_values(by = 'eu_sales', ascending = False)[:5].reset_index()
top_5_platforms_eu = top_5_platforms_eu['platform']


# JP region
top_5_platforms_jp = jp_customer[jp_customer['year_of_release'] >= 2013].pivot_table(index = 'platform', 
                      values = 'jp_sales', 
                      aggfunc = 'sum').sort_values(by = 'jp_sales', ascending = False)[:5].reset_index()
top_5_platforms_jp = top_5_platforms_jp['platform']


# OTHER region
top_5_platforms_other = other_customer[other_customer['year_of_release'] >= 2013].pivot_table(index = 'platform', 
                      values = 'other_sales', 
                      aggfunc = 'sum').sort_values(by = 'other_sales', ascending = False)[:5].reset_index()
top_5_platforms_other = top_5_platforms_other['platform']
```

```python
# –û—Ç—Ñ–∏–ª—å—Ç—Ä—É–µ–º –∏ —Å–≥—Ä—É–ø–ø–∏—Ä—É–µ–º –¥–∞–Ω–Ω—ã–µ –ø–æ –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã–º —Ä–µ–≥–∏–æ–Ω–∞–º

# NORTH AMERICA region
platform_sales_na = na_customer[(na_customer['platform'].isin(top_5_platforms_na) & (na_customer['year_of_release'] >= 2013) & (na_customer['year_of_release'] != 2016))].pivot_table(index = ['platform'],
                                                                               
                                                                               values = 'na_sales', 
                                                                               aggfunc = 'sum').reset_index().sort_values(by = 'na_sales', ascending = False)

# EUROPE region
platform_sales_eu = eu_customer[(eu_customer['platform'].isin(top_5_platforms_eu) & (eu_customer['year_of_release'] >= 2013) & (eu_customer['year_of_release'] != 2016))].pivot_table(index = ['platform'],
                                                                               
                                                                               values = 'eu_sales', 
                                                                               aggfunc = 'sum').reset_index().sort_values(by = 'eu_sales', ascending = False)

# JAPAN region
platform_sales_jp = jp_customer[(jp_customer['platform'].isin(top_5_platforms_jp) & (jp_customer['year_of_release'] >= 2013) & (jp_customer['year_of_release'] != 2016))].pivot_table(index = ['platform'],
                                                                               
                                                                               values = 'jp_sales', 
                                                                               aggfunc = 'sum').reset_index().sort_values(by = 'jp_sales', ascending = False)

# OTHER region
platform_sales_other = other_customer[(other_customer['platform'].isin(top_5_platforms_other) & (other_customer['year_of_release'] >= 2013) & (other_customer['year_of_release'] != 2016))].pivot_table(index = ['platform'],
                                                                               
                                                                               values = 'other_sales', 
                                                                               aggfunc = 'sum').reset_index().sort_values(by = 'other_sales', ascending = False)
```
